Below is an approximation of the best decision boundary from part 1 imposed on the a1_test.csv data 

<img src="2.1.png" alt="2.1" width="400"/>

For this confusion matrix A = True and B = False

| Model | A      | B      |
|-------|--------|--------|
| A     | TP = 2 | FP = 1 |
| B     | FN = 0 | TN = 1 |

Accuracy is defined as $\frac{(TP + TN)}{TP+FP+FN+TN}$

So the accuracy is $\frac{2+1}{2+1+1}$ = $\frac{3}{4}$ = 0.75%
